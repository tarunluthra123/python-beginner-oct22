### Print all digits (in reverse order)
Given N, print all of its digits.

**Example**: \
792

**Result**: \
2 9 7

````python
N = 548

while N > 0:
    r = N % 10
    print(r)
    N = N // 10
````

### Print N*M stars
Given integer N, M, print N*M star grid

**N = 3, M = 4** 
```
****
****
****
```

**N = 5, M = 5**
```
*****
*****
*****
*****
*****
```

````python
N = 10
M = 6

# Print N rows in total
for i in range(N):
    # For each row, we want to print M stars
    for j in range(M):
        print("*", end='')
        
    # Print a new line character
    print()
````

### Stair Pattern
Given N, print the following pattern stair pattern

**N = 5**
```
*
**
***
****
*****
```

````python
N = 3

for i in range(1, N+1):
    # Row no = i
    # Print i stars
    for j in range(i):
        print("*", end='')
    print()
````
